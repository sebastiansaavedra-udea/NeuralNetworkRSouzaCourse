## Formulacion matematica del problema

Es posible encontrar una funcion $\hat f(x) = \hat y$ que ajuste nuestros datos?

**El enfoque más sencillo en este caso es un modelo de regresión lineal**.

El modelo de regresión lineal simple es simplemente una función lineal de la característica de entrada. Con este enfoque se **asume** que existe una relación lineal entre las variables de entrada y las de salida.

Para la i-ésima instancia, el modelo de regresión lineal simple puede definirse como 

\begin{equation}
 \hat y^{(i)} = w_0x_0^{(i)} + w_1x_1^{(i)} 
\end{equation}

Donde $w_0$ and $w_1$ son los ***parametros*** del modelo (tambien llamados pesos) 

De forma más general, un modelo lineal realiza una predicción simplemente calculando una suma ponderada de las características de entrada, más una constante llamada término de sesgo (también llamado término de intercepción).
término)

<p><img alt="Colaboratory logo" height="300px" src="https://i.imgur.com/U3GzGtz.png" align="center" hspace="10px" vspace="0px"></p> 

\\



Podemos escribir el modelo de regrssion lineal simple de forma matricial. para esto debemos realizar las sigueintes definiciones:
\begin{equation}
W  = \begin{bmatrix}
                w_0 \\
                w_1 \\
                \end{bmatrix}
                , \ 
                 \textbf{x}^{(i)}  = \begin{bmatrix}
                x_0^{(i)}\\
                x_1^{(i)}\\
                \end{bmatrix}
                , \ 
                X  = \begin{bmatrix}
                \textbf{x}^{(0)T}\\
                \textbf{x}^{(1)T}\\
                \textbf{x}^{(2)T}\\
                .\\
                .\\
                \textbf{x}^{(N)T}\\
                \end{bmatrix}
                , \ 
                \hat Y  = \begin{bmatrix}
                \hat y^{(1)}\\
                \hat y^{(2)}\\
                \hat y^{(3)}\\
                .\\
                .\\
                \hat y^{(N)}\\
                \end{bmatrix}
\end{equation}

Donde:
* $N$ denota el numero total de instancias (filas) en el dataset
* $m$ denota el numero total de caracteristicas (columnas) en el dataset
* $x_0^{(i)} = 1$ para todo $i$ $\in$  N

De esta manera podemos escribir la regresion linal simple de manera matricial: 
\begin{equation}
 \hat y^{(i)} = W^T \textbf{x}^{(i)} 
\end{equation}


\begin{equation}
\hat Y = XW
\end{equation}









## Entrenamiento y funcion de costo 

Ahora tenemos que entrenar el modelo, lo que significa establecer los parámetros del modelo para que éste se ajuste lo mejor posible al conjunto de datos de entrenamiento. Primero necesitamos una medida de lo bien (o mal) que se ajusta el modelo a los datos. Así que podemos definir una función (a veces llamada función de perdida) que mida lo mal que lo hace el modelo. La medida de rendimiento más común de un modelo de regresión es el **Error Medio de Escuadrado**

$$Loss = (\hat y^{(i)} -  y^{(i)})^2 $$

Podemos calcular la funcion de perdida para todas nuestras instancias y luego hallar el promedio para obtener la funcion de costo

\begin{equation}
Cost = J = \frac{1}{N} \sum_{i=1}^N (\hat y^{(i)} -  y^{(i)} )^2
\end{equation}

\\

Como $J$ mide lo mal que le va a nuestro modelo, queremos minimizarlo. Específicamente tenemos que encontrar los valores de $W$ que minimizan la función de costo.


\begin{equation}
\underset{w_0, w_1}{argmin}\{J\}
\end{equation}


Podemos encontrar el valor de W que minimiza la función de pérdida tomando la derivada de $J$ con respecto a W y poniéndolas a cero. 


\begin{equation}
\nabla_{W} J  = \frac{2}{N} X^T(X W - Y) = 0
\end{equation}

Lo que da lugar a la llamada ecuación normal


\begin{equation}
W = (X^TX)^{-1} X^T Y
\end{equation}

# Metricas

Las métricas de rendimiento forman parte de todos los procesos de aprendizaje automático. Te dicen si estás progresando, y le ponen un número. Todos los modelos de Machine Learning necesitan una métrica para juzgar el rendimiento.

Las métricas son diferentes de las funciones de costo. Si bien Las funciones de costo muestran una medida del rendimiento del modelo, se utilizan principalmente para entrenar un modelo de machine learning (utilizando algún tipo de optimización), y suelen ser diferenciables en los parámetros del modelo. 

Las métricas se utilizan para controlar y medir el rendimiento de un modelo (durante el entrenamiento y las pruebas), y no necesitan ser diferenciables. 

Sin embargo, si para algunas tareas la métrica de rendimiento es diferenciable, también puede utilizarse como función de pérdida, como el MSE

## Root Mean Squared Error

El error cuadrático medio (RMSE) es la desviación estándar de los residuos (errores de predicción). Los residuos son una medida de la distancia a la que se encuentran los puntos de datos de la línea de regresión; el RMSE es una medida de la dispersión de estos residuos. En otras palabras, indica la concentración de los datos en torno a la línea de mejor ajuste.

\begin{equation}
RMSE = \sqrt{ \frac{1}{N} \sum_{i=1}^N (\hat y^{(i)} -  y^{(i)} )^2}
\end{equation}

\begin{equation}
R^2 = 1 - \frac{\sum_i (y_i - \hat y_i)^2}{\sum_i (y_i - \overline y)^2}
\end{equation}
